```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Define nodes and conditional probabilities
nodes = ["A", "B", "C"]
conditional_probs = {
    ("A", "B"): 0.7,  # P(B|A)
    ("A", "C"): 0.3,  # P(C|A)
}

# Add nodes to the graph
G.add_nodes_from(nodes)

# Add edges with conditional probabilities
for (parent, child), prob in conditional_probs.items():
    G.add_edge(parent, child, weight=prob)

# Specify node positions (optional)
pos = {"A": (0, 0), "B": (1, 1), "C": (1, -1)}

# Draw the graph
nx.draw(G, pos, with_labels=True, node_size=5000, node_color="skyblue", font_size=12, font_weight="bold")

# Add edge labels (conditional probabilities)
edge_labels = {(parent, child): f"P({child}|{parent}) = {prob}" for (parent, child), prob in conditional_probs.items()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10)

# Set plot title
plt.title("Conditional Probability Diagram")

# Display the plot
plt.show()
```
